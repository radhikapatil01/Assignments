nba dataset

```python
#importing the csv file
df1= pd.read_csv("/content/drive/MyDrive/nba.csv - nba.csv.csv")
df1
```

Replacing the categorical values to numerical values

```python
df1['Position'].replace(['SG','PF','PG','SF','C'],[0,1,2,3,4], inplace= True)
df1
```

Using label encoder

```python
#importing label encoder
from sklearn import preprocessing

label_encoder= preprocessing.LabelEncoder()

#encoding labels in column position
df1['Position']=label_encoder.fit_transform(df1['Position'])
df1['Position'].unique()
```

Quantative data into categorical data using pandas cut

```python
#adding a new column category for the age group

category=pd.cut(df1.Age,bins=[19,25,30,35,45], labels=['A','B','C','D'])
df1.insert(5,'Age_Group',category)
```
