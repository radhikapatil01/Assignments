Importing the necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

Loading the dataset using the seaborn library

```python
df=sns.load_dataset("titanic")
```

```python
df
```

```python
df.shape
```

Checking null or NaN values

```python
df.isnull().sum()
```

Dropping the NaN containing columns

```python
df.drop('deck', inplace=True, axis=1)
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df.drop('embark_town', inplace=True, axis=1)
```
